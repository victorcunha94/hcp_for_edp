```python
import numpy as np
import matplotlib.pyplot as plt

U = np.load("solution.npy")

plt.imshow(U.T, origin="lower", cmap="viridis")
plt.colorbar(label="Solution")
plt.title("Solucao por Jacobi Paralelizado")
plt.show()
```

```python

def uexact(x, y):
    return np.sin(np.pi * x) * np.sin(2*np.pi * y)
```

```python
#Residual test
N = U.shape[0]
dx = 1.0 / (N - 1)
x = np.linspace(0, 1, N)
y = np.linspace(0, 1, N)

X, Y = np.meshgrid(x, y, indexing='ij')
U_exact = uexact(X, Y)
residual = np.max(np.abs(U - U_exact))
print(f"Residuo maximo: {residual}")
```
